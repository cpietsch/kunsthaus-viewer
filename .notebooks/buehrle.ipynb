```python
import pandas as pd
from bs4 import BeautifulSoup

baseUrl = "https://www.buehrle.ch/"

with open('buehrle.html', 'r') as f:
    soup = BeautifulSoup(f, 'html.parser')

werk_items = soup.select('.werk')

def extractId(link):
    # example fullLink https://www.buehrle.ch/sammlung/?tx_buehrlewerkliste_werkliste%5Bwerk%5D=415&tx_buehrlewerkliste_werkliste%5Bcontext%5D=t&tx_buehrlewerkliste_werkliste%5Baction%5D=show&cHash=1b86b60e753a114c957d30cfb6483c32
    # extract 415
    return link.split('=')[1].split('&')[0]

data = []
for item in werk_items:
    link = item.select_one('a').get('href')
    fullLink = baseUrl + link
    uid = extractId(link)
    
    # first p is the artist, second p is the title
    artist = item.select_one('p').text.strip()
    title = item.select_one('p + p').text.strip()
    
    data.append([uid, fullLink, artist, title])

df = pd.DataFrame(data, columns=['uid', 'blink', 'artist', 'title'])
df.to_csv('buehrleLinks.csv', index=False)
df
```
